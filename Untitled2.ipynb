```python
out_path = 'obj-base.jsonl'
with open(out_path, 'w') as out:
  for i, datum in enumerate(data):
    with Image.open(datum['hypo_path']) as hypo_image:
      inputs = feature_extractor(images=hypo_image, return_tensors="pt")
      outputs = model(**inputs)
      logits = outputs.logits
      objects = list(filter(lambda n: n != 91, softmax(logits[0, :], dim=-1).max(dim=-1).indices.tolist()))
      datum['objects'] = objects
    out.write(f'{json.dumps(datum)}\n')
    if i % 10 == 0:
      print(i)
```
